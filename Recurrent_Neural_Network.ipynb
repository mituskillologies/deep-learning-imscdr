```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
list(enumerate(alphabets))
```

```python
# Create a dict with char as key and index as value
char_to_int = dict((c, i) for i, c in enumerate(alphabets))
```

```python
# Create a dict with index as key and char as value
int_to_char = dict((i, c) for i, c in enumerate(alphabets))
```

```python
int_to_char
```

```python
seq_length = 3
```

```python
# Prepare the data
x = []
y = []
for i in range(len(alphabets) - seq_length):
  # Previous character as input
  seq_in = alphabets[i:i + seq_length]
  # Next character as output
  seq_out = alphabets[i + seq_length]
  # Append the index of input value
  x.append([char_to_int[char] for char in seq_in])
  # Append the index of output value
  y.append(char_to_int[seq_out])
  # Print your input and output
  print(seq_in,'->',seq_out)
```

```python
y
```

```python
# Check the data after preparation
import pandas as pd
pd.DataFrame({
    'Input': x,
    'Output': y
})
```

```python
# Reshape the data
import numpy as np
```

```python
x = np.reshape(x, (len(x), seq_length, 1))
```

```python
x.shape
```

```python
# Feature scaling
x = x / 25
```

```python
x
```

```python
# Create the output in one-hot encoded format
from keras.utils import to_categorical
```

```python
y = to_categorical(y)
```

```python
y
```

```python
x.shape
```

```python
y.shape
```

```python
# Import the RNN classes
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
# Create the RNN model
model = Sequential()

# Add the input layer
model.add(LSTM(32, input_shape=(x.shape[1], x.shape[2])))

# Add output layer
model.add(Dense(26, activation='softmax'))
```

```python
# Get the summary structure of the model
model.summary()
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the algorithm
model.fit(x, y, epochs=1000, batch_size=1)
```

```python
# Evaluate the model
model.evaluate(x, y, batch_size=1)
```

```python
# Prediction on new data
new = 'K'
# Extract the integer value of char
new = char_to_int[new]
# Reshape it to 1 x 1
new = np.reshape(new, (1, 1, 1))
# Scaling the feature
new = new / 25
# Predict on the new data
pred = model.predict(new, verbose=0)
# Find the highest probability
np.argmax(pred)
# Extract the character
print('Next Character:',int_to_char[np.argmax(pred)])
```
